```python
# -----------------------------
# Student Management Console App
# -----------------------------

students = []   # List of dictionaries


# Function: Add new student
def add_student():
    print("\n--- Add New Student ---")
    name = input("Enter name: ").strip()

    # Validate age (no negative, must be number)
    while True:
        try:
            age = int(input("Enter age: "))
            if age < 0:
                print("Age cannot be negative! Try again.")
            else:
                break
        except ValueError:
            print("Invalid input! Age must be a number.")

    # Validate marks
    while True:
        try:
            marks = float(input("Enter marks: "))
            if marks < 0:
                print("Marks cannot be negative! Try again.")
            else:
                break
        except ValueError:
            print("Invalid input! Marks must be a number.")

    # Store in dictionary
    students.append({"name": name, "age": age, "marks": marks})
    print(f"\nâœ” Student '{name}' added successfully!\n")


# Function: View all students
def view_students():
    print("\n--- All Students ---")

    if not students:
        print("No student records found.\n")
        return

    print("{:<20} {:<10} {:<10}".format("Name", "Age", "Marks"))
    print("-" * 45)

    for s in students:
        print("{:<20} {:<10} {:<10}".format(s["name"], s["age"], s["marks"]))

    print()


# Function: Search a student by name
def search_student():
    print("\n--- Search Student ---")
    name = input("Enter name to search: ").strip()

    found = False
    for s in students:
        if s["name"].lower() == name.lower():
            print("\nStudent Found:")
            print(f"Name  : {s['name']}")
            print(f"Age   : {s['age']}")
            print(f"Marks : {s['marks']}\n")
            found = True
            break

    if not found:
        print(f"\nâŒ No student found with name '{name}'\n")


# Function: Update marks
def update_marks():
    print("\n--- Update Student Marks ---")
    name = input("Enter name of student: ").strip()

    for s in students:
        if s["name"].lower() == name.lower():

            while True:
                try:
                    new_marks = float(input("Enter new marks: "))
                    if new_marks < 0:
                        print("Marks cannot be negative! Try again.")
                    else:
                        break
                except ValueError:
                    print("Invalid input! Enter numbers only.")

            s["marks"] = new_marks
            print(f"\nâœ” Marks updated for '{name}'!\n")
            return

    print(f"\nâŒ Student '{name}' not found!\n")


# Main menu loop
def main_menu():
    while True:
        print("========== Student Management Console ==========")
        print("A. Add New Student")
        print("B. View All Students")
        print("C. Search Student by Name")
        print("D. Update Student Marks")
        print("E. Exit")
        print("================================================")

        choice = input("Choose an option (A/B/C/D/E): ").strip().upper()

        if choice == "A":
            add_student()
        elif choice == "B":
            view_students()
        elif choice == "C":
            search_student()
        elif choice == "D":
            update_marks()
        elif choice == "E":
            print("\nExiting program... Goodbye! ðŸ‘‹")
            break
        else:
            print("\nInvalid option! Please choose A, B, C, D, or E.\n")


# Run the program
main_menu()
```
